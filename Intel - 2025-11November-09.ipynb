# Scoping



```python
x = 100

for i in range(5):
    x = i

print(x)    
```

```python
x = 100

def myfunc():
    print(f'x = {x}')  # local x? NO. global x? YES, 100

print(f'Before, x = {x}')
myfunc()
print(f'After, x = {x}')
```

# Scoping : L E G B

- Local -- start here if we're in a function
- (Enclosing)
- Global -- start here if we're outside of a function
- Builtin

```python
'x' in globals()
```

```python
myfunc.__code__.co_varnames     # local variables known to the function
```

```python
x = 100

def myfunc():
    x = 200
    print(f'x = {x}')

print(f'Before, x = {x}')
myfunc()
print(f'After, x = {x}')
```
