# Scoping



```python
x = 100

for i in range(5):
    x = i

print(x)    
```

```python
x = 100

def myfunc():
    print(f'x = {x}')  # local x? NO. global x? YES, 100

print(f'Before, x = {x}')
myfunc()
print(f'After, x = {x}')
```

# Scoping : L E G B

- Local -- start here if we're in a function
- (Enclosing)
- Global -- start here if we're outside of a function
- Builtin

```python
'x' in globals()
```

```python
myfunc.__code__.co_varnames     # local variables known to the function
```

```python
x = 100

def myfunc():
    x = 200
    print(f'x = {x}')  # local x? YES, 200

print(f'Before, x = {x}')  # global x? YES, 100
myfunc()  
print(f'After, x = {x}')  # global x? YES, 100
```

```python
myfunc.__code__.co_varnames
```

```python
x = 100

def myfunc():
    print(f'x = {x}')  # local x? YES..... 
    x = 200  # hoisting problem

print(f'Before, x = {x}') 
myfunc()  
print(f'After, x = {x}')  
```

```python
myfunc.__code__.co_varnames
```

```python
x = 100

def myfunc():
    x +=1    # x = x + 1
    print(f'x = {x}') 

print(f'Before, x = {x}') 
myfunc()  
print(f'After, x = {x}')  
```

```python
x = 100

def myfunc():
    x = 200
    print(f'x = {x}') 

print(f'Before, x = {x}') 
myfunc()  
print(f'After, x = {x}')  
```

```python
# I want to change the global x!

x = 100

def myfunc():
    global x   
    x = 200
    print(f'x = {x}') 

print(f'Before, x = {x}') 
myfunc()  
print(f'After, x = {x}')  
```

```python
myfunc.__code__.co_varnames
```

```python
dis.dis(myfunc)
```

```python
y = [10, 20, 30]

def myfunc():
    y[0] = '!'     # y.__setitem__(0, '!')
    print(f'y = {y}') 

print(f'Before, y = {y}') 
myfunc()  
print(f'After, y = {y}')  
```

```python
y = [10, 20, 30]

def myfunc():
    y = [10, 20, 30]
    y[0] = '!'     # y.__setitem__(0, '!')
    print(f'y = {y}') 

print(f'Before, y = {y}') 
myfunc()  
print(f'After, y = {y}')  
```
