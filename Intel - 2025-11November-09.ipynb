# Scoping : L E G B

- Local -- start here if we're in a function
- (Enclosing)
- Global -- start here if we're outside of a function
- Builtin

```python
'x' in globals()
```

```python
myfunc.__code__.co_varnames     # local variables known to the function
```

```python
x = 100

def myfunc():
    x = 200
    print(f'x = {x}')  # local x? YES, 200

print(f'Before, x = {x}')  # global x? YES, 100
myfunc()  
print(f'After, x = {x}')  # global x? YES, 100
```

```python
myfunc.__code__.co_varnames
```

```python
x = 100

def myfunc():
    print(f'x = {x}')  # local x? YES..... 
    x = 200  # hoisting problem

print(f'Before, x = {x}') 
myfunc()  
print(f'After, x = {x}')  
```

```python
myfunc.__code__.co_varnames
```

# Comprehensions



```python
numbers = list(range(10))
numbers
```

```python
# I want these numbers to the 2nd power, and in a new list

output = []

for one_number in numbers:
    output.append(one_number ** 2)

output    
```

```python
# list comprehension

# - if I have an iterable value
# - I want a list based on it
# - I have a Python expression to translate from the first to the second

[one_number ** 2
 for one_number in numbers]
```

```python
text = 'this is a test sentence'

# I want every word to start with a capital letter

text.title()  # assume this doesn't exist... what then?
```

```python
' '.join([one_word.capitalize()
         for one_word in text.split()])
```

```python
[one_number ** 2              # SELECT  -- expression
 for one_number in numbers]   # FROM    -- iteration
```

# Exercise

1. Ask the user to enter a string containing numbers separated by spaces. Add these numbers together, and print their sum. There is a builtin `sum` function that takes a list of numbers and returns the total.
2. Ask the user to enter text. Print the total number of characters *except* for spaces. Use a comprehension to do this!

```python
words = input('Enter numbers: ').split()

# - I have: a list of strings, each string is only digits 0-9
# - I want: a list of ints
# - I can translate from the first to the second with "int"

sum([int(one_item)
    for one_item in words])
```

```python
words = input('Enter numbers: ').split()

# - I have: a list of strings, words
# - I want: a list of ints, the length of each word
# - Translate with: len

sum([len(one_word)
    for one_word in words])
```

```python
for one_line in open('/etc/passwd'):
    print(one_line.split(':')[0])
```

```python
# usernames from /etc/passwd

[one_line.split(':')[0]                # SELECT
 for one_line in open('/etc/passwd')   # FROM
 if not one_line.startswith('#')]      # WHERE
```
