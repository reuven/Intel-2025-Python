# Comprehensions



```python
numbers = list(range(10))
numbers
```

```python
# I want these numbers to the 2nd power, and in a new list

output = []

for one_number in numbers:
    output.append(one_number ** 2)

output    
```

```python
# list comprehension

# - if I have an iterable value
# - I want a list based on it
# - I have a Python expression to translate from the first to the second

[one_number ** 2
 for one_number in numbers]
```

```python
def line_to_dict(one_line):
    brand, color, size = one_line.strip().split('\t')

    return {'brand': brand,
           'color':color,
           'size':size}

shoes = [line_to_dict(one_line)
            for one_line in open('shoe-data.txt')]
```

```python
import operator
sorted(shoes, key=operator.itemgetter('size'))   # ['size']
```
